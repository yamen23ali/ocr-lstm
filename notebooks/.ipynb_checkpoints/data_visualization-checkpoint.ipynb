## Show samples from the dataset

```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def get_samples(book_dir,images_num=9):
    images = []
    texts= []
    
    image_files = glob.glob(book_dir + "/*.png")
    chosen_imgs = np.random.choice(image_files, images_num, replace=False)
    for img_file in chosen_imgs:
        images.append(img_file)
        
        base_name = os.path.splitext(os.path.splitext(os.path.basename(img_file))[0])[0]
        text_file = f'{book_dir}/{base_name}.gt.txt' # get text file name
        with open(text_file,encoding = 'utf-8') as f:
            texts.append(f.read().rstrip())
                
    return images, texts
            
def show(subcorpora_dir, images_num=9):
    images, texts = get_samples(subcorpora_dir, images_num)
    
    fig = plt.figure(figsize=(12, 12))  # figure size in inches
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    
    for i in range(0, images_num):
        fig = plt.figure(111)
        a = fig.add_subplot(2, 1, i + 1, xticks=[], yticks=[])
        img = mpimg.imread(images[i])
        imgplot = plt.imshow(img, cmap=plt.cm.binary, interpolation='nearest')
        a.set_title(str(texts[i]))


show('../GT4HistOCR/corpus/RefCorpus-ENHG-Incunabula/1476-Historij-Wierstaat/',2)    
```

# Get Min and Max sample from the dataset

```python
def get_min_max_image_size(dataset_src_dir):
    max_size=(0,0)
    max_img_file = ''

    min_size=(1000000,1000000)
    min_img_file = ''

    corporas_dirs = [x[0] for x in os.walk(dataset_src_dir)][1:]
    for corporas_dir in corporas_dirs:
        books_dirs = [x[0] for x in os.walk(corporas_dir)][1:]
        for book_dir in books_dirs:
            image_files = glob.glob(book_dir + "/*.png")
            for img_file in image_files:
                with Image.open(img_file) as img:
                    img_size = img.size

                    if max_size < img_size:
                        max_size = img_size
                        max_img_file = img_file

                    if min_size > img_size :
                        min_size = img_size
                        min_img_file = img_file
                        
    return min_size,min_img_file, max_size, max_img_file

def get_basename(file):
    return os.path.splitext(os.path.splitext(os.path.basename(file))[0])[0]

```

```python
dataset_src_dir='../dataset/'
min_size,min_img_file, max_size, max_img_file = get_min_max_image_size(dataset_src_dir)
```

```python

print('Max Image size:',max_size)
print('Image', max_img_file)
print('\n')
print('in ætrurĩam & inde ad urbis romæ partes eſt reductus. Secundus rex alfonſus: ſuum')
img = mpimg.imread(max_img_file)
imgplot = plt.imshow(img, cmap=plt.cm.binary, interpolation='nearest')
```

## Compiling several text files into a single CSV file

```python
import csv
from pathlib import Path

data_dir = '../dataset/EarlyModernLatin/1471-Orthographia-Tortellius'
all_orthographia_tortellius = '1471-Orthographia-Tortellius.csv'
with open(all_orthographia_tortellius, 'w', encoding='Latin-1') as out_file:
    csv_out = csv.writer(out_file)
    csv_out.writerow(['content'])
    for fileName in Path(data_dir).glob('*.txt'):
        lines = [ ]
        with open(str(fileName.absolute()),'rb') as one_text:
            for line in one_text.readlines():
                lines.append(line.decode(encoding='Latin-1',errors='ignore').strip())
        csv_out.writerow([' '.join(lines)])
```

```python
train = pd.read_csv(all_orthographia_tortellius)
train.head(5)
```

### 1.1 Number of Words


```python
train['word_count'] = train['content'].apply(lambda x: len(str(x).split(" ")))
train[['content','word_count']].head(5)
```

```python
plt.figure(figsize=(50,30))
plt.title("Word Frequencies in " + all_orthographia_tortellius)
plt.ylabel("Total Number of Occurrences")
plt.scatter(train['content'], train['word_count'],s=100)
plt.show()
```

### 1.2 Number of characters

```python
train['char_count'] = train['content'].str.len() ## this also includes spaces
train[['content','char_count']].head(5)
```

```python
plt.figure(figsize=(50,30))
plt.title("Chars Frequencies in " + all_orthographia_tortellius)
plt.ylabel("Total Number of Occurrences")
plt.scatter(train['content'], train['char_count'],s=70)
plt.show()
```

### 1.3 Average Word Length

```python
def avg_word(sentence):
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))

train['avg_word'] = train['content'].apply(lambda x: avg_word(x))
train[['content','avg_word']].head(5)
```

### 1.4 Number of stopwords


```python
from nltk.corpus import stopwords
stop = stopwords.words('german')

train['stopwords'] = train['content'].apply(lambda x: len([x for x in x.split() if x in stop]))
train[['content','stopwords']].head(5)


```

```python
print('All stop words', stop)

train['content'].apply(lambda x: [print(x) for x in x.split() if x in stop])
```
